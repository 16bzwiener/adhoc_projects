This function will drop every row that has any null values unless you pass in the columns you want to check.
```python
ss = ss.fillna(0)
ss.count()
```

```python
url = 'https://raw.githubusercontent.com/jerrytigerxu/NBA-Salary-Prediction/master/data/nba_salaries.csv'

salaries = pd.read_csv(url)
salaries = salaries.drop(salaries.columns[0], axis=1)

salaries.head()
```

```python
# creating a player leverage column
salaries['Player Leverage'] = salaries['Salary'] / salaries['Team Payroll']

# creating a league weight column
salaries['League Weight'] = salaries['Salary'] / salaries['Total NBA Payroll']

# Creating a Team Market Size column
salaries['Team Market Size'] = salaries['Team Payroll'] / salaries['Total NBA Payroll']

salaries.head()
```

```python
for i in salaries['Year'].unique():
    #highest = salaries[salaries['Year'] == i]['League Weight']
    print(salaries[salaries.Year == i][salaries[salaries.Year == i]['League Weight'] == salaries
                                       [salaries.Year == i]['League Weight'].max()])
    

```

```python
salaries.loc[salaries.groupby('Year')['League Weight'].idxmax()]
```

```python
# time to group teams into regions
salaries['US Region'] = " "

northeast = ['New York Knicks', 'Boston Celtic', 'Brooklyn Nets', 'Philadelphia 76ers', 'Toronto Raptors']
midwest = ['Cleveland Caveliers', 'Detroit Pistons', 'Indiana Pacers', 'Milwaukee Bucks', 'Minnesota Timberwolves', 'Chicago Bulls']
south = ['Atlanta Hawks', 'Charlotte Hornets', 'Miami Heat', 'Orlando Magic', 'New Orleans Pelicans', 'Washington Wizards', 'Dallas Mavericks', 'Oklahoma City Thunder', 'Houston Rockets', 'San Antonio Spurs', 'Memphis Grizzlies']
west = ['Golden State Warriors', 'Denver Nuggets', 'Utah Jazz', 'Sacramento Kings', 'Los Angeles Lakers', 'Los Angeles Clippers', 'Phoenix Suns', 'Portland Trail Blazers']

region = {}
for team in northeast:
    region[team] = 'Northeast'
for team in midwest:
    region[team] = 'Midwest'
for team in south:
    region[team] = 'South'
for team in west:
    region[team] = 'West'
    
for i in salaries.index:
    salaries.at[i, 'US Region'] = region[salaries.at[i, 'Team']]
    
salaries.sample(5)
```

```python
playerstats = pd.merge(ss, salaries, 
                                how='left', left_on=['Year', 'Player'], right_on=['Year', 'Player'])

playerstats.head()
```

```python
playerstats = playerstats[playerstats.Year > 1990]
playerstats.count()
```

```python
from sklearn.model_selection import train_test_split
```

```python
colnames = list(playerstats) 

colnames.remove('Player')
colnames.remove('Pos')
colnames.remove('US Region')
colnames.remove('Salary')
colnames.remove('League Weight')
colnames.remove('Total NBA Payroll')
colnames.remove('Team Payroll')
colnames.remove('Team')
colnames.remove('Player Leverage')
print(colnames)
```

```python
y = playerstats['League Weight']
X = playerstats[colnames]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print('Coefficients: \n', lm.coef_)
```

```python
import numpy as np

highest = np.argmax(lm.coef_)
value = np.amax(lm.coef_)
print(value)


var = colnames[highest]
var
```

```python
predictions = lm.predict(X_test)
```

```python
# Here we'll plot our predictions versus the actual values

plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics


print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.tree import DecisionTreeRegressor

dtree = DecisionTreeRegressor()
dtree.fit(X_train, y_train)
```

```python
dt_predictions = dtree.predict(X_test)

print('MAE:', metrics.mean_absolute_error(y_test, dt_predictions))
print('MSE:', metrics.mean_squared_error(y_test, dt_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, dt_predictions)))
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
rfr_predictions = rfr.predict(X_test)
print('MAE:', metrics.mean_absolute_error(y_test, rfr_predictions))
print('MSE:', metrics.mean_squared_error(y_test, rfr_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rfr_predictions)))
```
